```python
import numpy as np
import matplotlib.pyplot as plt

# Input range
x = np.linspace(-10, 10, 100)

# 1. Binary Step
def binary_step(x):
    return np.where(x >= 0, 1, 0)

# 2. Linear
def linear(x):
    return x

# 3. ReLU
def relu(x):
    return np.maximum(0, x)

# 4. Leaky ReLU
def leaky_relu(x, alpha=0.01):
    return np.where(x >= 0, x, alpha * x)

# 5. Softmax
def softmax(x):
    exp_x = np.exp(x - np.max(x))  # stability trick
    return exp_x / exp_x.sum()

# Compute outputs
y_step = binary_step(x)
y_linear = linear(x)
y_relu = relu(x)
y_leaky = leaky_relu(x)
y_softmax = softmax(x)

# Plot graphs
plt.figure(figsize=(10,6))

plt.plot(x, y_step, label="Binary Step")
plt.plot(x, y_linear, label="Linear")
plt.plot(x, y_relu, label="ReLU")
plt.plot(x, y_leaky, label="Leaky ReLU")
plt.plot(x, y_softmax, label="Softmax")

plt.title("Activation Functions")
plt.xlabel("Input")
plt.ylabel("Output")
plt.legend()
plt.grid()

plt.show()
```
